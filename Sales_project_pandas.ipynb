# HOW TO MERGE ALL CSV FILES INTO 1 FILE

```python
df = pd.DataFrame([])
for root, dirs, files in os.walk("."):
   for name in files:
       df_temp = pd.read_csv(name)
       df = pd.concat([df,df_temp])
df.to_csv('all-csv-files.csv')    
```

```python
df.head()
```

# REMOVE UN-NAMED COLUMNS

```python
df.drop(df.columns[df.columns.str.contains('Unnamed',case = False)],axis = 1, inplace = True)
```

```python
df
```

# How to find Nan Values


```python
df2= df[df.isna().any(axis=1)]
df2.head()
```

```python
df=df.dropna(how="all")
df.head()
```

# Remove ' Or'

```python
df2=df[df["Order Date"].str[0:2] == "Or" ]
df2.head()
```

```python
df=df[df["Order Date"].str[0:2] != "Or"]
df.head()
```

```python
df["Month"]=df["Order Date"].str[0:2]
df["Month"]=df["Month"].astype("int32")
df.head()
```

# Create Sales Column And convert columns into numeric data type

```python
df['Quantity Ordered']=pd.to_numeric(df["Quantity Ordered"])
df['Price Each']= pd.to_numeric(df["Price Each"])
df.head()
```

```python
df["Sales"]=df["Quantity Ordered"] * df['Price Each']
df.head()
```



# Highest Sales Month 

```python
Results=df.groupby(["Month"]).sum()
Results
```

```python
import matplotlib.pyplot as plt
months= range(1,13)
plt.bar(months,Results["Sales"])


plt.xlabel("MONTHS")
plt.ylabel("Sales in US($)")
plt.show()
```

# City Highest Sale And split Column

```python
def get_city(address):
    return address.split(',')[1]

def get_sate(address):
    return address.split(',')[2].split(' ')[1]



df["Cities"]=df["Purchase Address"].apply(lambda x: f"{get_city(x)} ({get_sate(x)})")

df.head()
```

```python
Results=df.groupby(["Cities"]).sum()
Results
```

# Find BEst Sale Time of Customers

```python
df['Order Date']= pd.to_datetime(df["Order Date"])
```

```python
df.head()
```

```python
df['Hours']=df['Order Date'].dt.hour
df['Minutes']=df['Order Date'].dt.minute
df.head()
```

```python
import matplotlib.pyplot as plt
hour=[hour for hour, df in df.groupby("Hours")]
plt.plot=(hour, df.groupby(["Hours"]).count())


plt.show()
```

# Products OFten Sold Together (Duplicated)

```python
df=df[df["Order ID"].duplicated(keep=False)]

df["Group_Product"]=df.groupby("Order ID")["Product"].transform(lambda x: ','.join(x))
df.head(20)
```

```python
df=df[["Order ID","Group_Product"]].drop_duplicates()
df.head()
```

# MOst Selling Product

```python
product_group=df.groupby("Product")

quantity_ordered= product_group.sum()["Quantity Ordered"]

products=[product for product, df in product_group ]

plt.bar(products,quantity_ordered)

plt.xticks(products, rotation="vertical", size=8)
```

```python
prices= df.groupby('Product').mean()['Price Each']
prices
```

```python
fig, ax1 = plt.subplots()

ax2 = ax1.twinx()
ax1.bar(products, quantity_ordered,color='g' )
ax2.plot(products, prices, 'b-')

ax1.set_xlabel('Products_Name')
ax1.set_ylabel('quantity_ordered', color='g')
ax2.set_ylabel('prices', color='b')
ax1.set_xticklabels(products, rotation='vertical', size=8)

plt.show()
```
